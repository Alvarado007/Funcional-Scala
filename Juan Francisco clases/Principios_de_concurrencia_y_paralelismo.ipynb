```scala
def Log1(msg:String):Unit =
    println(s"${Thread.currentThread().getName}: $msg")

def thread(cuerpo: => Unit): Unit = {
    val t = new Thread {
        override def run(): Unit = cuerpo
    }
    t.start()
    t
}
```

```scala
var uidCount = 0
def getUniqueUID():Long = {
    val freshUid = uidCount + 1
    Thread.sleep(1000)
    uidCount = freshUid
    freshUid
}
```

```scala
def printUniqueIds(n:Int):Unit ={
    val uids = for (i <- 0 until n) yield getUniqueUID()
    Log1(s"uids: $uids")
}
printUniqueIds(5)
```
